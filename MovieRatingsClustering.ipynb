# Introduction

The goal of this project is to use clustering techniques to classify movie-watchers into groups using how they rate movies of various genres. This will allow us to further understand the customer base and could be useful for further classification and research. Our data was pulled from the MovieLens user rating data set.

```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

Our first data set has ratings for different movies organized by user. We will be combining this data set with another to extract the genre information for each movie.

```python
print('There are {} total entries in the dataset, with {} unique users and {} unique movies.'.format(len(ratings['userId']),ratings['userId'].nunique(),ratings['movieId'].nunique()))
```

All columns are of the expected type and there doesn't appear to be any missing values.

```python
ratings['rating'].value_counts()
```

There doesn't appear to be any unexpected ratings value, with 10 different scores ranging from 0.5 to 5. 

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

The movies dataframe shows the title of each movie and which genre they are in. We are going to merge this data set on the movieId column to get the genre and rating information into one dataframe.

```python
movies.info()
```

```python
#Merging the movie and ratings dataset
movieratings_df = pd.merge(movies,ratings,on='movieId')
movieratings_df.head()
```

Merging the two dataframes we now have a single dataframe with all information contained. The dataframe has each movieId starting from 1 and each user rating and their corresponding ID. The genre and title information for the movie is also contained.

```python
movieratings_df.info()
```

```python
movieratings_df['genres'].value_counts()
```

```python

#Create columns for each genre
chosen = ['Comedy','Romance','Horror','Action']
for name in chosen:
    movieratings_df[name] = np.nan

#Label each genre column as 1 if the movie is that genre, otherwise 0
for name in chosen:
    movieratings_df.loc[movieratings_df['genres'].str.contains(name), name] = 1

movieratings_df = movieratings_df.fillna(0)
```

```python

# Explore the counts of genres in our dataset.
Comedy_mov = 0
Action_mov = 0
Romance_mov = 0
Horror_mov = 0
for movie in movieratings_df['Comedy']:
    if movie == 1:
       Comedy_mov += 1
for movie in movieratings_df['Horror']:
    if movie == 1:
       Horror_mov += 1
for movie in movieratings_df['Action']:
    if movie == 1:
       Action_mov += 1
for movie in movieratings_df['Romance']:
    if movie == 1:
       Romance_mov += 1

print('There are {} Horror, {} Action, {} Romance, and {} Comedy movies in our dataset.'.format(Horror_mov,Action_mov,Romance_mov,Comedy_mov))
```

```python

# Create a Genre Dataframe that shows Average genre rating for each User
genre_df = pd.DataFrame()

com_movies = movieratings_df[movieratings_df['Comedy']==1]
avg_com_votes_per_user = com_movies[com_movies['movieId'].isin(com_movies['movieId'])].loc[:, ['userId', 'rating']].groupby(['userId'])['rating'].mean().round(2)
genre_df = pd.concat([genre_df,avg_com_votes_per_user],axis=1)
rom_movies = movieratings_df[movieratings_df['Romance']==1]
avg_rom_votes_per_user = rom_movies[rom_movies['movieId'].isin(rom_movies['movieId'])].loc[:, ['userId', 'rating']].groupby(['userId'])['rating'].mean().round(2)
genre_df = pd.concat([genre_df,avg_rom_votes_per_user],axis=1)
act_movies = movieratings_df[movieratings_df['Action']==1]
avg_act_votes_per_user = act_movies[act_movies['movieId'].isin(act_movies['movieId'])].loc[:, ['userId', 'rating']].groupby(['userId'])['rating'].mean().round(2)
genre_df = pd.concat([genre_df,avg_act_votes_per_user],axis=1)
hor_movies = movieratings_df[movieratings_df['Horror']==1]
avg_hor_votes_per_user = hor_movies[hor_movies['movieId'].isin(hor_movies['movieId'])].loc[:, ['userId', 'rating']].groupby(['userId'])['rating'].mean().round(2)
genre_df = pd.concat([genre_df,avg_hor_votes_per_user],axis=1)

genre_df.columns = ['Avg_Comedy','Avg_Romance','Avg_Action','Avg_Horror']
genre_df.index.name = 'UserID'
genre_df.head()
```

```python
genre_df.info()
```

```python
genre_df = genre_df.dropna()
```

```python
# Let's plot the data to compare users interest in Comedy vs Horror, and Action vs Romance.
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
a = genre_df['Avg_Comedy']
b = genre_df['Avg_Horror']
sns.scatterplot(a,b,alpha=0.75)

plt.subplot(1,2,2)
c = genre_df['Avg_Action']
d = genre_df['Avg_Romance']
sns.scatterplot(c,d,alpha=0.75)
plt.show()
```

Now let's implement k-means clustering to see if it can be used as a way to interpret user preference.



```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0)

X_1 = genre_df[['Avg_Comedy','Avg_Horror']].values
X_2 = genre_df[['Avg_Action','Avg_Romance']].values

y_pred = kmeans.fit_predict(X_1)
z_pred = kmeans.fit_predict(X_2)

#Plot the clustered solution using K=2
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
sns.scatterplot(a,b,hue=y_pred)

plt.subplot(1,2,2)
sns.scatterplot(c,d,hue=z_pred)
plt.show()
```

As you can see there is a clear distinction between the two clusters in our genre plots. The Comedy vs Horror plot shows that the clusters split around where the Avg_Horror rating equals about 3. The Action vs Romance cutoff isn't quite as flat, occuring somewhere between Action equalling 3 to 3.5. But how do we select K?



```python
# Selecting k - The Elbow Method
from scipy.spatial.distance import cdist
distortions_1 = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X_1)
    kmeanModel.fit(X_1)
    distortions_1.append(sum(np.min(cdist(X_1, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X_1.shape[0])

distortions_2 = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X_2)
    kmeanModel.fit(X_2)
    distortions_2.append(sum(np.min(cdist(X_2, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X_2.shape[0])
    

# Plot the elbow
plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.plot(K, distortions_1, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k - Horror vs. Comedy')

plt.subplot(1,2,2)
plt.plot(K, distortions_2, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k - Action vs Romance')
plt.show()
```

Both the plots above aid us in choosing the optimal k for our k-means algorithms. It appears as if 3 to 5 are good options for the Horror vs Comedy plot, and good choices for the Action vs Romance plot are 3 and 6. Let's try again with the optimal k parameters.



```python
from sklearn.cluster import KMeans
kmeans_1 = KMeans(n_clusters=4, random_state=0)
kmeans_2 = KMeans(n_clusters=6, random_state=0)

X_1 = genre_df[['Avg_Comedy','Avg_Horror']].values
X_2 = genre_df[['Avg_Action','Avg_Romance']].values

y_pred = kmeans_1.fit_predict(X_1)
z_pred = kmeans_2.fit_predict(X_2)

#Plot the clustered solution
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
sns.scatterplot(a,b,hue=y_pred)

plt.subplot(1,2,2)
sns.scatterplot(c,d,hue=z_pred)
plt.show()
```

As you can see, the users preferences in each plot seem to become more clear when using the optimal k values. I believe this could be a very successful way in predicting the preferences of specific users and therefore further understanding the behavior of users.



```python

#Evaluating each solution using Silhoutte Coefficient
from sklearn import metrics
from sklearn.metrics import pairwise_distances

model_1 = metrics.silhouette_score(X_1, y_pred, metric='euclidean')
model_2 = metrics.silhouette_score(X_2, z_pred, metric='euclidean')

print('The silhoutte coefficient for Comedy vs Horror is {}.'.format(model_1))
print('The silhoutte coefficient for Action vs Romance is {}.'.format(model_2))
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

#Model 1
bandwidth_1 = estimate_bandwidth(X_1, quantile=0.25, n_samples=len(X_1))
ms = MeanShift(bandwidth=bandwidth_1, bin_seeding=True)
ms.fit(X_1)
labels_1 = ms.labels_
n_clusters_1 = len(np.unique(labels_1))

#Model 2
bandwidth_2 = estimate_bandwidth(X_2, quantile=0.25, n_samples=len(X_2))
ms = MeanShift(bandwidth=bandwidth_2, bin_seeding=True)
ms.fit(X_2)
labels_2 = ms.labels_
n_clusters_2 = len(np.unique(labels_2))


print("First Solution estimated clusters: {}".format(n_clusters_1))
print("Second Solution estimated clusters: {}".format(n_clusters_2))
```

```python
#Plot the clustered solution using mean shift labels
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
sns.scatterplot(a,b,hue=labels_1)

plt.subplot(1,2,2)
sns.scatterplot(c,d,hue=labels_2)
plt.show()
```

Similar number of clusters are predicted as is with k-means, although the performance appears visually quite different with 1 or 2 clusters having the majority of the data points in the set. Let's use the silhoutte coefficient to look at quantifying the performance.



```python

#Evaluating each solution using Silhoutte Coefficient

from sklearn import metrics
from sklearn.metrics import pairwise_distances
model_3 = metrics.silhouette_score(X_1, labels_1, metric='euclidean')
model_4 = metrics.silhouette_score(X_2, labels_2, metric='euclidean')
print('The silhoutte coefficient for Mean-Shift Comedy vs Horror is {}.'.format(model_3))
print('The silhoutte coefficient for Mean-Shift Action vs Romance is {}.'.format(model_4))
```

Looking at the silhoutte coefficient it appears the Mean-Shift algorithm is about 0.1 higher than the k-means approach.



```python
#DBSCAN Solution
from sklearn.cluster import DBSCAN 
db_default_1 = DBSCAN(eps = 0.0875, min_samples = 10).fit(X_1) 
labels_3 = db_default_1.labels_ 
db_default_2 = DBSCAN(eps = 0.0875, min_samples = 10).fit(X_2) 
labels_4 = db_default_2.labels_

#Plot the DBScan clustered solution
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
sns.scatterplot(a,b,hue=labels_3)

plt.subplot(1,2,2)
sns.scatterplot(c,d,hue=labels_4)
plt.show()
```

```python

#Evaluating each solution using Silhoutte Coefficient

from sklearn import metrics
from sklearn.metrics import pairwise_distances
model_5 = metrics.silhouette_score(X_1, labels_3, metric='euclidean')
model_6 = metrics.silhouette_score(X_2, labels_4, metric='euclidean')
print('The silhoutte coefficient for Mean-Shift Comedy vs Horror is {}.'.format(model_5))
print('The silhoutte coefficient for Mean-Shift Action vs Romance is {}.'.format(model_6))
```

Negative silhouette coefficients are indicating that this appears to not be a good solution for this scenario and dataset.
